# IMPORTING REQUIRED LIBRARIES:

```python
import pandas as pd
from sklearn import preprocessing
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
data = pd.read_csv('hr.csv')
```

```python
data.head()
```

# REMOVING NON-ESSENTIAL FEATURES:

```python
data = data.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

```python
data1 = data.copy()
```

```python
data.head()
```

# CONVERTING CATEGORICAL DATA INTO NUMERIC DATA:

As the ML Algorithms do not interpret character values, we have to preprocess them into numeric values by using:

1. Label Encoding
2. One Hot Encoding

Both the models are being trained simentaneously, so dont get confused.
- Learn these technique to kick off.

```python
# LABEL ENCODING
label_encoder = preprocessing.LabelEncoder() 
# Encode labels in column 'species'. 
data['Attrition']= label_encoder.fit_transform(data['Attrition'])
data['BusinessTravel']= label_encoder.fit_transform(data['BusinessTravel']) 
data['Department']= label_encoder.fit_transform(data['Department']) 
data['EducationField']= label_encoder.fit_transform(data['EducationField']) 
data['Gender']= label_encoder.fit_transform(data['Gender']) 
data['JobRole']= label_encoder.fit_transform(data['JobRole']) 
data['MaritalStatus']= label_encoder.fit_transform(data['MaritalStatus']) 
data['OverTime']= label_encoder.fit_transform(data['OverTime']) 
data.head()
```

```python
# ONE HOT ENCODING
df = data1
list  = ['BusinessTravel', 'Department',
       'DistanceFromHome', 'Education', 'EducationField',
       'EnvironmentSatisfaction', 'Gender', 'JobInvolvement',
       'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus',
       'NumCompaniesWorked', 'OverTime',
        'PerformanceRating', 'RelationshipSatisfaction',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
       'YearsSinceLastPromotion', 'YearsWithCurrManager']
for i in list:
    df = pd.concat([df,pd.get_dummies(data1[i], prefix=i,drop_first=True)],axis=1)
    # now drop the original 'country' column (you don't need it anymore)
    df.drop([i],axis=1, inplace=True)
df['Attrition']= label_encoder.fit_transform(df['Attrition'])
```

```python
data1.head()
```

NOTE: Making copies of your dataset is very helpful in feature selection and engineering.

```python
data1 = df.copy()
```

# RESMAPLING THE DATA TO REDUCE CLASS IMBALANCE:

- Suppose your target variable has two classes in it, 0 and 1.
- Your ML ALgorithm will predict whether the class for the given data is 0 or 1.
- You have to create the final dataset which will be having equal no. of datapoints of class as 0 and 1. 
- Suppose you use 20 datapoints having class 0, than you have to use 20 datapoints having class as 1 also.
- Not doing so will make the algorithm bias towards one class.

```python
data_0 = data[data['Attrition']==0]
data_1 = data[data['Attrition']==1]
data1_0 = data1[data1['Attrition']==0]
data1_1 = data1[data1['Attrition']==1]
```

```python
use_0 = data_0.sample(474)
data_use = pd.concat([use_0,data_1,data_1],axis=0)
use1_0 = data1_0.sample(474)
data1_use = pd.concat([use1_0,data1_1,data1_1],axis=0)
```

- Now the model will be trained on this dataset, "data_use".
- The parent dataset has 237 rows with class 1 and 1233 rows with class 0.
- We will double the class 1 datapoints and only select 474 datapoints from class 0.
- The new dataset created has 474 rows with class 1 and 474 rows with class 0.

# DIVIDING THE DATASET INTO TRAINING-TESTING DATASET:

We'll first store all the features in X, and target variable in y.

```python
y_le = data_use['Attrition']
X_le = data_use.drop('Attrition',axis=1)

y_ohe = data1_use['Attrition']
X_ohe = data1_use.drop('Attrition',axis=1)
```

Train_test_split will divide the dataset into training data (80%) and testing data (20%)

```python
from sklearn.model_selection import train_test_split
X_train_le,X_test_le,Y_train_le,Y_test_le = train_test_split(X_le,y_le,test_size = 0.20,random_state =117)
X_train_ohe,X_test_ohe,Y_train_ohe,Y_test_ohe = train_test_split(X_ohe,y_ohe,test_size = 0.2,random_state =117)
```

# MODELLING

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import xgboost as xgboost
```

- Use the official XGBoost documentation for better understading of modelling.
- TWo models are getting trained here.One is getting trained on label encoded data, the other one's getting trained on one hot encoded data.

```python
modelXg_le = xgboost.XGBClassifier()
modelXg_ohe = xgboost.XGBClassifier()

modelXg_le.fit(X_train_le,Y_train_le)
pred_le = modelXg_le.predict(X_test_le)

modelXg_ohe.fit(X_train_ohe,Y_train_ohe)
pred_ohe = modelXg_ohe.predict(X_test_ohe)
```

# PERFORMANCE EVALUATION OF THE MODEL: 

```python
print('XGBOOST PERFORMANCE FOR LABEL ENCODING')
print('--------------------------------------------------')
print(classification_report(Y_test_le,pred_le))
print('Training Accuracy: ',accuracy_score(Y_train_le,modelXg_le.predict(X_train_le)))
print('Testing Accuracy: ',accuracy_score(Y_test_le,pred_le))
print('Confusion Matrix:\n',confusion_matrix(Y_test_le,pred_le))
print('--------------------------------------------------')
print('XGBOOST PERFORMANCE FOR ONE-HOT ENCODING')
print('--------------------------------------------------')
print(classification_report(Y_test_ohe,pred_ohe))
print('Training Accuracy: ',accuracy_score(Y_train_ohe,modelXg_ohe.predict(X_train_ohe)))
print('Testing Accuracy: ',accuracy_score(Y_test_ohe,pred_ohe))
print('Confusion Matrix:\n',confusion_matrix(Y_test_ohe,pred_ohe))
print('--------------------------------------------------')
```
